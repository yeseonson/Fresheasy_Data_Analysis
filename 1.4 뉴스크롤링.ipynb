# 코로나 외식

```python
import requests
from bs4 import BeautifulSoup
```

```python
url_format = 'https://search.naver.com/search.naver?&where=news&query={0}&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=3&ds={1}&de={2}&docid=&nso=so:r,p:from{3}to{4},a:all&mynews=0&cluster_rank=34&start={5}'
```

```python
r = requests.get(url_format.format('코로나 외식', '2020.01.01', '2020.12.20', '20200101', '20201220', '1'))
```

```python
sp = BeautifulSoup(r.text, 'html.parser')
```

```python
sources = sp.select('div.group_news > ul.list_news > li div.news_area > a')
```

```python
len(sources)
```

```python
for source in sources:
    print(source.attrs['title'])
```

```python
SEARCH_WORD = '코로나 외식'
START_DATE = '2020.01.01'
END_DATE = '2020.12.20'
START_DATE_2 = '20200101'
END_DATE_2 = '20201220'
```

```python
title = []
# 페이지 크롤링
for page_number in range(100):
    r = requests.get(url_format.format(SEARCH_WORD, START_DATE, END_DATE, START_DATE_2, END_DATE_2, str(page_number*10+1)))
    sp = BeautifulSoup(r.text, 'html.parser')
    sources = sp.select('div.group_news > ul.list_news > li div.news_area > a')
    
    # 뉴스 기사 크롤링
    for source in sources:
        news_r = source.get_text()
        title.append(news_r)
        
title
```

```python
len(title)
```

```python
data = ','.join(title)
data
```

```python
from konlpy.tag import Twitter; t = Twitter()
tokens_ko = t.nouns(data)
```

```python
import nltk
ko = nltk.Text(tokens_ko, name = '코로나 외식')
ko.vocab()
```

```python
import platform
import matplotlib.pyplot as plt
%matplotlib inline

from bs4 import BeautifulSoup 
from wordcloud import WordCloud, STOPWORDS
import nltk
from fbprophet import Prophet
from datetime import datetime


path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')   
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['코로나','거리','시대','외식','업계','곳','로','업','중','명','최','수','속','두기','비대']

ko = [each_word for each_word in ko if each_word not in stop_words]
```

```python
ko = nltk.Text(ko, name='코로나 외식')
plt.figure(figsize=(12,6))
ko.plot(50)    
plt.show()
```

```python
data1 = ko.vocab().most_common(150)
wordcloud = WordCloud(font_path='/Library/Fonts/malgun.ttf',
                      relative_scaling = 0.1,
                      background_color='white',
                      ).generate_from_frequencies(dict(data1))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
