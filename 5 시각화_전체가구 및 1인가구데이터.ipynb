```python
pop1 = pd.DataFrame(
    data = {'연도' : ['2015','2016','2017','2018','2019'],
            '총가구':[19111030, 19367696, 19673875, 19979188, 20343188],
            '1인가구': [5203440, 5397615, 5618677, 5848594, 6147516]},
                 columns=['연도','총가구','1인가구'])
pop1
```

```python
pop1['1인가구 비율'] = round(pop1['1인가구']/pop1['총가구']*100,2)
pop1
```

```python
pop1['1인가구 비율'].plot(color = '#17becf',marker='o', linestyle='--');
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5)
years = ['2015', '2016', '2017','2018','2019']
values = [27.23, 27.87, 28.56, 29.27, 30.22]

plt.bar(x, values, color='yellowgreen',width=0.5)
plt.xticks(x, years)
plt.ylim([25, 31])
plt.xlabel('연도')
plt.ylabel('1인가구 비율')
plt.show()
```

```python
x = np.arange(5)
years = ['2015', '2016', '2017','2018','2019']
values = [27.23, 27.87, 28.56, 29.27, 30.22]

plt.figure(figsize=(12,6))
plt.bar(x, values, color='yellowgreen',width=0.5)
plt.xticks(x, years)
plt.ylim([25, 31])
plt.xlabel('연도')
plt.ylabel('1인가구 비율')
plt.title('연도별 1인가구 비율')

for i, v in enumerate(x):
    plt.text(v, values[i], values[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 12, 
             color='black',
             horizontalalignment='center',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)

plt.show()
```
